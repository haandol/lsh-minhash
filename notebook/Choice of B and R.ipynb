```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
rows = 200
```

```python
def f(t, b, r):
    return 1-(1-t**r)**b
```

```python
t_range = np.arange(0, 1, 0.001)

data = []
for t in t_range:
    for b in range(1, rows//2+1):
        r = rows // b
        if b * r != rows:
            continue
        data.append({'b': b, 't': 1 if 1 < t else t, 'val': f(t, b, r)})
```

```python
data[:5]
```

```python
data[-5:]
```

```python
reference = []
for x in np.arange(0, 1, 0.001):
    reference.append({'b': 'ref', 't': x, 'val': (x>0.17)*1})
```

```python
reference[:5]
```

```python
reference[-5:]
```

```python
data.extend(reference)
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 4))
sns.lineplot(x='t', y='val', hue='b', data=df)
```
